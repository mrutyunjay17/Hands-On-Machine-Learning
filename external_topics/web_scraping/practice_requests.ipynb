## ITunes Search

```python
base_site = 'https://itunes.apple.com/search'
```

```python
url = base_site + "?term=the+beatles%country=us"
requests.get(url)
```

```python
#Rewriting
r = requests.get(base_site, params={"term":"the beatles", "country":"us"})
```

```python
r.status_code
```

```python
r.url
```

```python
info = r.json()
print(json.dumps(info, indent=3))
```

```python
info.keys()
```

```python
print(json.dumps(info['results'][0], indent=4))
```

```python
info['resultCount']
```

```python
r = requests.get(base_site, params={"term":"the beatles","country":"us", "limit":4})
```

```python
r.json()['resultCount']
```

```python
r = requests.get(base_site, params={"media":"hahaha", "term":"alternative","country":"us"})
```

```python
r.status_code, r.json()
```

## Auth

```python
import requests
import json
import pandas as pd
```

```python
APP_ID = "~get api id~"
APP_KEY = "get api key"
```

```python
api_endpoint = "https://api.edamam.com/api/nutrition-details"

url = api_endpoint + "?app_id="+APP_ID+"&app_key="+APP_KEY
```

```python
headers = {
    "Content-Type":"application/json"
}

recipe = {
    'title': 'Cappuccino',
    'ingr': ['18g ground espresso', '150ml milk']
}
```

```python
r = requests.post(url, headers=headers, json=recipe)
r.status_code
```

```python
capp_info = r.json()
print(json.dumps(capp_info, indent=4))
```

```python
capp_info['totalNutrients']['SUGAR']
```

```python
pd.DataFrame(capp_info['totalNutrients'])
```

```python
capp_nutrients = pd.DataFrame(capp_info['totalNutrients']).transpose()
capp_nutrients 
```

```python
capp_nutrients.to_csv("Cappuccino_nutrients.csv")
```
